```python
## Import necessary libraries
import kagglehub
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# 1Ô∏è‚É£ Descargar dataset
path = kagglehub.dataset_download("amruthayenikonda/dirty-dataset-to-practice-data-cleaning")
print("Path to dataset files:", path)
```

```python
# 2Ô∏è‚É£ Ver archivos del folder
print(os.listdir(path))
```

```python
# 3Ô∏è‚É£ Cargar CSV correcto
csv_path = os.path.join(path, "my_file (1).csv")
df = pd.read_csv(csv_path)
print(df.info())
```

```python
# 4Ô∏è‚É£ Limpiar nombres de columnas
df.columns = df.columns.str.replace(r'[^\x00-\x7F]+', '', regex=True).str.strip()
print("Columnas limpias:", df.columns.tolist())
```

```python
# 5Ô∏è‚É£ Renombrar columnas dif√≠cil de usar
df.rename(columns={
    "Actualgross": "Actual gross",
    "Adjustedgross (in 2022 dollars)": "Adjusted gross"
}, inplace=True)
```

```python
# 6Ô∏è‚É£ Limpiar valores num√©ricos (EL PASO M√ÅS IMPORTANTE)
cols_to_clean = ["Actual gross", "Adjusted gross", "Average gross"]

for col in cols_to_clean:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(r'[^\d.]', '', regex=True)  # Remove $, commas, [b], [e], letters, symbols
        .replace('', np.nan)                     # Replace empty strings with NaN
        .astype(float)                           # Convert to float
    )
```

```python
# 7Ô∏è‚É£ Manejar filas vac√≠as / columnas con demasiados nulos
df = df.dropna(axis=1, thresh=int(0.7 * len(df)))
```

```python
# 8Ô∏è‚É£ Quitar columna in√∫til
if "Ref." in df.columns:
    df = df.drop(columns="Ref.")
```

```python
# 9Ô∏è‚É£ Arreglo manual del "Rank" incorrecto
df.at[7, "Rank"] = 8
```

```python
# üîü Poner Rank como √≠ndice
df.set_index("Rank", inplace=True)
```

```python
# 11 Ordenar por Actual gross
df_sorted_by_Agross = df.sort_values(by="Actual gross", ascending=False)
df['Tour title'] = df['Tour title'].str.replace('‚Ä†','')
df['Tour title'] = df['Tour title'].str.replace('‚Ä°[4][a]','')
df['Tour title'] = df['Tour title'].str.replace('*','')
df['Tour title'] = df['Tour title'].str.replace('‚Ä°[21][a]','')
```
